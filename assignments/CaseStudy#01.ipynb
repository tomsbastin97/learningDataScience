```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append('$PATH/python3.8/site-packages')
import seaborn as sns
```

### 1. Read the dataset to the python environment.

```python
dataset = pd.read_csv('cars_data.csv')
dataset
```

### 2. Check for the null values present in the dataset.

```python
print('Number of null values under each column of the dataset:')
print(dataset.isnull().sum())
```

### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
female = dataset[dataset['Buyer Gender']=='Female']
fCount = len(female.index)
male = dataset[dataset['Buyer Gender']=='Male']
mCount = len(male.index)

nData = {'Male': mCount, 'Female': fCount}
plt.figure(figsize=(10,5))
plt.barh(list(nData.keys()), list(nData.values()))
for index, value in enumerate(list(nData.values())):
    plt.text(value, index, str(value))

```

### 4. Find the top 5 cars based on their sales price.

```python
expensive = dataset.sort_values(by=['Sale Price'])
expensive[['Make', 'Model', 'Sale Price']].tail(5)
```

### 5. Find the least 5 cars based on their Resell price.

```python
cheap = dataset.sort_values(by=['Resell Price'])
cheap[['Make', 'Model', 'Resell Price']].head(5)
```
