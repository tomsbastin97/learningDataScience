### 2. The company needs to check whether there is any dependency between the features “Region” and “Manager”.

H0: 'Region' and 'Manager' are probably independent variables

H1: 'Region' and 'Manager' are probably dependent variables

```python
dataset['Manager'].unique()
```

```python
dataset['Region'].unique()
```

```python
man_num={'Manager - A':1,'Manager - B':2,'Manager - C':3}
reg_num={'Region - A':1,'Region - B':2,'Region - C':3}
```

```python
Manager_Num = [man_num[item] for item in dataset.Manager]
Region_Num = [reg_num[item] for item in dataset.Region]
```

```python
dataset['Manager_Num']=Manager_Num
dataset['Region_Num']=Region_Num
dataset
```

```python
from scipy.stats import chi2_contingency
table = dataset.drop(['Month','Manager','Region','Sales_before_digital_add(in $)','Sales_After_digital_add(in $)'],axis=1)
stat, p, dof, expected = chi2_contingency(table)
print('stat=%.3f, p=%.3f' % (stat, p))
if p < 0.05:
    print('The variables Region and Manager are probably dependent')
else:
    print('The variables Region and Manager are probably independent')
```
