# Case Study #05 - ANOVA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

### Read the dataset to the python environment.

```python
dataset = pd.read_csv('W9Sales_data_zone_wise.csv')
dataset
```

### 1. The mean sales generated by each zone.

```python
pd.set_option('precision', 3)
print('Zone A:', dataset['Zone - A'].mean().__round__(3))
print('Zone B:', dataset['Zone - B'].mean().__round__(3))
print('Zone C:', dataset['Zone - C'].mean().__round__(3))
print('Zone D:', dataset['Zone - D'].mean().__round__(3))
```

### 2. Total sales generated by all the zones for each month.

```python
dataset['Total Sales'] = dataset.sum(axis=1, numeric_only=True)
dataset
```

### 3. Check whether all the zones generate the same amount of sales.

Lets define hypothesis as follows:\
H0 = All zones generate same amount of sales\
H1 = Atleast one zone is different from others sales\
alpha = 0.05

```python
f_stats,p_value = stats.f_oneway(dataset['Zone - A'],dataset['Zone - B'],dataset['Zone - C'],dataset['Zone - C'])
print(f_stats,p_value)
```

```python
if (p_value<=0.05):
    print(p_value,"","Since p value is less than 0.05 we reject the null hypothesis","\n Thus all zones have difference in the sales")
else:
    print(p_value,"","Since p value is greater than 0.05 we accept the null hypothesis")
```
